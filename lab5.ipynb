```python
! pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def step(x):
    return np.where(x > 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    e_x = np.exp(x - np.max(x))  # stability trick
    return e_x / e_x.sum()


# Input range
x = np.linspace(-10, 10, 400)

# Plot functions
plt.figure(figsize=(12, 8))

activations = {
    "Step": step,
    "Sigmoid": sigmoid,
    "Tanh": tanh,
    "ReLU": relu,
    "Leaky ReLU": leaky_relu
}

for i, (name, func) in enumerate(activations.items(), 1):
    plt.subplot(2, 3, i)
    y = func(x)
    plt.plot(x, y)
    plt.title(name)
    plt.grid(True)

# Separate Softmax demo (works on vectors, not single x)
plt.subplot(2, 3, 6)
sample_x = np.array([1, 2, 3])
y = softmax(sample_x)
plt.bar(range(len(sample_x)), y)
plt.title("Softmax (Example: [1,2,3])")
plt.grid(True)

plt.tight_layout()
plt.show()
```
